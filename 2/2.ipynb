```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.boxplot(y= df['Second year:   Sem 1'])
```

```python
sem_col = ['First year:   Sem 1','First year:   Sem 2','Second year:   Sem 1','Second year:   Sem 2']
for col in sem_col:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lowerBound = q1 -1.5*iqr
    outerBound = q3 +1.5*iqr

    outliers = df[(df[col]< lowerBound) | (df[col]>outerBound)]
    display(outliers)

    median = df[col].median()
    df[col]= np.where((df[col]< lowerBound) | (df[col]>outerBound),median,df[col])
```

```python


# Log transformation (adding small constant to avoid log(0))
df['Log_SecondYear_Sem1'] = np.log(df['Second year:   Sem 1'] )

# Plot original vs transformed
```

```python
plt.figure(figsize=(12, 5))

```

```python
plt.subplot(1, 2, 1)
sns.histplot(df['Second year:   Sem 1'], kde=True)
plt.title("Original Distribution")

```

```python
plt.subplot(1, 2, 2)
sns.histplot(df['Log_SecondYear_Sem1'], kde=True)
plt.title("Log-Transformed Distribution")

```

```python
plt.tight_layout()
plt.show()
```
