```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("autodata.csv")
```

```python
df.head(10) 
```

```python
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['stroke'].head()
```

```python
modeStroke = df['stroke'].mode().astype('float').iloc[0]
```

```python
df['stroke']=df['stroke'].fillna(modeStroke)
df['stroke'].isnull().sum()
```

```python
df['horsepower'].head(5)
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].mean())
df['horsepower'].isnull().sum()
```

```python
df['peak-rpm'].head(15)
```

```python
bins = np.linspace(df['peak-rpm'].min(),df['peak-rpm'].max(),4)
groups = ['low', 'mid', 'high']
df['binnedPeakRPM'] = pd.cut(df['peak-rpm'],bins=bins, labels = groups, include_lowest = True)
df[['binnedPeakRPM','peak-rpm']].head(20)
```

```python
df['city-L/100km'] =235/df['city-mpg']
```

```python
df[['city-L/100km','city-mpg']].head(20)
```

```python
df['length'] = 10*df['length']
```

```python
df['length'].head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['scaled-lenght'] = scaler.fit_transform(df[['length']])
df[['scaled-lenght','length']].head()
```

```python
df['width'] = 10*df['width']
df['width'].head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['scaled-width'] = scaler.fit_transform(df[['width']])
df[['scaled-width','width']].head()
```

```python
df['transformedPrice'] = np.log(df['price']+1)
df[['transformedPrice','price']].head(10)
```

```python
plt.hist(df['transformedPrice'], bins=20, alpha=0.499, label='Transformed Price (Log)')
plt.legend()
plt.title('Histogram: Transformed Price')
plt.show()
plt.xscale('log')
plt.hist(df['price'], bins=10, alpha=0.7, label='Original Price')
plt.hist(df['transformedPrice'], bins=20, alpha=0.7, label='Transformed Price (Log)')
plt.legend()
plt.title('Histogram: Original vs. Transformed Price')
plt.show()

```
